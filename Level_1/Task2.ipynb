```python
import pandas as pd

df = pd.read_csv(r"D:\Cognifyz_Internship\Dataset\Dataset.csv")

print(df.head())
```

```python
print(df.describe())
```

```python
top_cities = df['City'].value_counts().head(10)

print(top_cities)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x=top_cities.index, y=top_cities.values)

plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)

plt.show()
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)

print(top_cuisines)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x=top_cuisines.values, y=top_cuisines.index)

plt.title("Top 10 Cuisines")
plt.xlabel("Count")
plt.ylabel("Cuisine")

plt.show()
```

```python
city_ratings = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

print(city_ratings)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x=city_ratings.index, y=city_ratings.values)

plt.title("Top Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```

## Conclusion

- Descriptive statistical analysis was performed on the restaurant dataset.
- Most restaurants are concentrated in a few major cities.
- Certain cuisines are significantly more popular than others.
- Some cities have higher average restaurant ratings.
- Visualizations helped identify important restaurant distribution patterns.
