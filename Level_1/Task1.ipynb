```python
import pandas as pd

df = pd.read_csv("../Dataset/Dataset .csv")

print(df.head())
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
# Check missing values
print(df.isnull().sum())

# Fill missing cuisine values
df['Cuisines'] = df['Cuisines'].fillna('Unknown')

# Verify again
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
print(df['Aggregate rating'].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(df['Aggregate rating'], bins=10, kde=True)

plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')

plt.show()
```

## Conclusion

- The dataset contains 9551 rows and 21 columns.
- Missing values were found only in the 'Cuisines' column.
- Missing values were handled successfully.
- The dataset contains numerical and categorical features.
- The average aggregate rating is approximately 2.66.
- Most restaurants have ratings between 3.0 and 4.5.
