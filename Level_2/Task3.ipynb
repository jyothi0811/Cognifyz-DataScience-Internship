```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Cognifyz_Internship\Dataset\Dataset.csv")

print(df.head())
```

```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].apply(len)

print(df[['Restaurant Name', 'Restaurant_Name_Length']].head())
```

```python
print(df['Restaurant_Name_Length'].describe())
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Restaurant_Name_Length'], bins=20)

plt.title("Restaurant Name Length Distribution")
plt.xlabel("Name Length")
plt.ylabel("Count")

plt.show()
```

```python
def rating_category(rating):
    if rating >= 4.0:
        return 'Excellent'
    elif rating >= 3.0:
        return 'Good'
    elif rating >= 2.0:
        return 'Average'
    else:
        return 'Poor'

df['Rating_Category'] = df['Aggregate rating'].apply(rating_category)

print(df[['Aggregate rating', 'Rating_Category']].head())
```

```python
print(df['Rating_Category'].value_counts())
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Rating_Category', data=df)

plt.title("Restaurant Rating Categories")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
correlation = df[
    ['Restaurant_Name_Length', 'Aggregate rating', 'Votes']
].corr()

print(correlation)
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.title("Feature Correlation Heatmap")

plt.show()
```

## Conclusion

- Feature engineering techniques were applied to the restaurant dataset.
- New features such as restaurant name length and rating category were created.
- Rating categories helped group restaurants based on quality.
- Correlation analysis showed relationships between engineered features and ratings.
- Feature engineering improved the dataset for future machine learning tasks.
