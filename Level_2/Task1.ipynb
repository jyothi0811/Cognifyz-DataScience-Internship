```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Cognifyz_Internship\Dataset\Dataset.csv")

print(df.head())
```

```python
table_booking_percentage = (
    df['Has Table booking']
    .value_counts(normalize=True) * 100
)

print(table_booking_percentage)
```

```python
online_delivery_percentage = (
    df['Has Online delivery']
    .value_counts(normalize=True) * 100
)

print(online_delivery_percentage)
```

```python
table_booking_ratings = (
    df.groupby('Has Table booking')['Aggregate rating']
    .mean()
)

print(table_booking_ratings)
```

```python
plt.figure(figsize=(6,5))

sns.barplot(
    x=table_booking_ratings.index,
    y=table_booking_ratings.values
)

plt.title("Average Ratings by Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Average Rating")

plt.show()
```

```python
price_delivery = pd.crosstab(
    df['Price range'],
    df['Has Online delivery']
)

print(price_delivery)
```

```python
price_delivery.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Online Delivery Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=0)

plt.show()
```

## Conclusion

- The percentage of restaurants offering table booking and online delivery was analyzed.
- Restaurants with table booking generally received higher average ratings.
- Online delivery availability varies across different price ranges.
- Visualization techniques helped compare restaurant services effectively.
- The analysis provided useful insights into customer convenience features.
