```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Cognifyz_Internship\Dataset\Dataset.csv")

print(df.head())
```

```python
price_distribution = df['Price range'].value_counts()

print(price_distribution)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Price range', data=df)

plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```

```python
average_rating = df.groupby('Price range')['Aggregate rating'].mean()

print(average_rating)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=average_rating.index,
    y=average_rating.values
)

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```

```python
delivery_price = pd.crosstab(
    df['Price range'],
    df['Has Online delivery']
)

print(delivery_price)
```

```python
delivery_price.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Online Delivery Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=0)

plt.show()
```

## Conclusion

- Restaurants are distributed across different price ranges.
- Some price ranges contain significantly more restaurants than others.
- Higher price ranges generally have slightly higher average ratings.
- Online delivery availability varies depending on price category.
- Visualization helped identify pricing and delivery trends effectively.
